## Decision Tree for Regression

Load Scikit-Learn diabetes dataset

```python
import pandas as pd
```

```python
# load diabetes dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target

# X_BMI  = X[:, 2]
# print(X_BMI)
# filter for the BMI feature
X_BMI = X[:,2].reshape(-1,1)
print(X_BMI)
```

```python
# X_BMI
# X_BMI= pd.DataFrame(X_BMI)
# X_BMI
```

Fit a Decision Tree model to the data

```python
from sklearn import tree

model = tree.DecisionTreeRegressor(max_depth=2)
model = model.fit(X_BMI, y)
```

Plot the Decision Tree together with the data

```python
# plot the data as a scatter plot, and the decision tree as a line
import matplotlib.pyplot as plt
plt.scatter(X_BMI, y, c="orange")

# create a range of values for the x-axis
import numpy as np
tree_x = np.linspace(min(X_BMI), max(X_BMI), 1000)
print(tree_x)

plt.plot(tree_x, model.predict(tree_x), c="black")

# add labels
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.title("Decision Tree Regression")

plt.show()
```

Plot a diagram of the Decision Tree

```python
tree.plot_tree(model)
```

## Decision Trees for Classification

```python
# load diabetes dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
X, y = diabetes.data, diabetes.target
model = tree.DecisionTreeRegressor(max_depth=2)
model = model.fit(X, y)

tree.plot_tree(model)
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
model = tree.DecisionTreeClassifier()
model = model.fit(X, y)
```

```python
tree.plot_tree(model)
```
